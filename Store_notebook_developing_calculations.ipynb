### Linear Regression Models for Sales and Profit Prediction

#### Data Analysis
We begin by analyzing our dataset, which comprises investment, sales, and profit data. Let's take a closer look at the key statistics:

- **Investment**:
  - Mean: [mean investment value]
  - Standard Deviation: [standard deviation of investment]

  - Standard Deviation: [standard deviation of profit]

#### Data Preparation
After analyzing our data, we split it into features (investment) and target variables (sales and profit). We then proceeded to split the data into training and testing sets.

#### Model Training
Two separate Linear Regression models were trained:
- One for predicting sales based on investment.
- Another for predicting profit based on investment.

#### Model Evaluation
We evaluated the performance of our models using Mean Squared Error (MSE) and Mean Absolute Error (MAE) metrics on the testing data.

- **Sales Model**:
  - MSE: [MSE value for sales prediction]
  - MAE: [MAE value for sales prediction]

- **Profit Model**:
  - MSE: [MSE value for profit prediction]
  - MAE: [MAE value for profit prediction]

#### Conclusion
Our Linear Regression models show promising results in predicting sales and profit based on investment. However, further analysis and refinement may be necessary to improve the accuracy and robustness of the predictions.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Data preparation
data = {
    'Investment': [55500, 55500, 55500, 55500, 55500],  # Assuming investment is constant for all data points
    'Sales': [38376, 2357.10, 0, 0, 0],  # Update with actual sales data
    'Profit': [-14766.90, 0, 0, 0, 0]  # Update with actual profit data
}

df = pd.DataFrame(data)

# Separate features (X) and target variables (y)
X = df[['Investment']]
y_sales = df['Sales']
y_profit = df['Profit']

# Split data into training and testing sets
X_train, X_test, y_train_sales, y_test_sales = train_test_split(X, y_sales, test_size=0.2, random_state=42)
_, _, y_train_profit, y_test_profit = train_test_split(X, y_profit, test_size=0.2, random_state=42)

# Train linear regression model for sales prediction
sales_model = LinearRegression()
sales_model.fit(X_train, y_train_sales)

# Train linear regression model for profit prediction
profit_model = LinearRegression()
profit_model.fit(X_train, y_train_profit)

# Make predictions
sales_predictions = sales_model.predict(X_test)
profit_predictions = profit_model.predict(X_test)

# Evaluate the models
mse_sales = mean_squared_error(y_test_sales, sales_predictions)
mae_sales = mean_absolute_error(y_test_sales, sales_predictions)
mse_profit = mean_squared_error(y_test_profit, profit_predictions)
mae_profit = mean_absolute_error(y_test_profit, profit_predictions)

print("Sales Model:")
print("MSE:", mse_sales)
print("MAE:", mae_sales)

print("\nProfit Model:")
print("MSE:", mse_profit)
print("MAE:", mae_profit)
```
