# Analyzing Annual Sales vs Investments using Deep Learning Model

This code snippet demonstrates the use of a deep learning model to analyze the relationship between annual sales and investments over a period of time.

## Sample Data Description

- **Years**: An array representing the years over the given period. For example, `[1, 2, 3, 4, 5]`.
- **Sales**: Total sales for each year. For example, `[10000, 20000, 30000, 40000, 50000]`.
- **Investments**: Total investments made for each year. For example, `[5000, 6000, 7000, 8000, 9000]`.

## Deep Learning Model

The deep learning model consists of three layers:
1. **Input Layer**: Dense layer with 64 units and ReLU activation function.
2. **Hidden Layer**: Dense layer with 64 units and ReLU activation function.
3. **Output Layer**: Dense layer with 1 unit.

The model is compiled using the Adam optimizer and mean squared error loss function.

## Training

The model is trained using the `fit()` function for 1000 epochs.

## Predictions

- Annual sales are predicted based on the provided investment data.
- Profit is calculated by subtracting investments from predicted sales.
- Sales are predicted for future years beyond the provided data.

## Visualization

- Actual sales and predicted sales are visualized using bar charts.
- Investments are plotted as a reference line.
- Profit is visualized as a bar chart.
- Predicted sales for future years are also shown.



````python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Sample data - yearly sales and investments
years = np.array([1, 2, 3, 4, 5])  # Years
sales = np.array([10000, 20000, 30000, 40000, 50000])  # Total sales for each year
investments = np.array([5000, 6000, 7000, 8000, 9000])  # Total investments for each year

# Define the deep learning model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=[1]),  # Input layer
    tf.keras.layers.Dense(units=64, activation='relu'),  # Hidden layer
    tf.keras.layers.Dense(units=1)  # Output layer
])

# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")

# Train the model
model.fit(years, sales, epochs=1000, verbose=0)  # Training the model on sales data

# Predict annual sales based on investments
predicted_sales = model.predict(years)

# Visualize the results
plt.figure(figsize=(10, 6))

# Plot actual sales and predicted sales
plt.bar(years - 0.2, sales, width=0.4, color='b', align='center', label='Actual Sales')
plt.bar(years + 0.2, predicted_sales.flatten(), width=0.4, color='r', align='center', label='Predicted Sales')

# Plot investments as a reference
plt.plot(years, investments, marker='o', linestyle='-', color='g', label='Investments')

# Predict sales for future years
future_years = np.array([6, 7, 8, 9, 10])
predicted_sales_future = model.predict(future_years)
plt.bar(future_years, predicted_sales_future.flatten(), width=0.4, color='m', align='center', label='Predicted Sales (Future)')

plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Annual Sales vs Investments')
plt.xticks(np.concatenate([years, future_years]))  # Include future years on x-axis
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
````

# Analyzing Annual Sales vs Investments using Deep Learning Model

This code snippet demonstrates the use of a deep learning model to analyze the relationship between annual sales and investments over a period of time.

## Objective

The objective of this analysis is to predict annual sales based on investment data using a deep learning model and visualize the results.

## Sample Data Description

- **Years**: An array representing the years over the given period. For example, `[1, 2, 3, 4, 5]`.
- **Sales**: Total sales for each year. For example, `[10000, 20000, 30000, 40000, 50000]`.
- **Investments**: Total investments made for each year. For example, `[5000, 6000, 7000, 8000, 9000]`.

## Deep Learning Model

The deep learning model consists of three layers:
1. **Input Layer**: Dense layer with 64 units and ReLU activation function.
2. **Hidden Layer**: Dense layer with 64 units and ReLU activation function.
3. **Output Layer**: Dense layer with 1 unit.

The model is compiled using the Adam optimizer and mean squared error loss function.

## Training

The model is trained using the `fit()` function for 1000 epochs with sales data as the target variable.

## Predictions

- Annual sales are predicted based on the provided investment data.
- Profit is calculated by subtracting investments from predicted sales.
- Sales are predicted for future years beyond the provided data.

## Visualization

- Actual sales and predicted sales are visualized using bar charts.
- Investments are plotted as a reference line.
- Profit is visualized as a bar chart.
- Predicted sales for future years are also shown.

This analysis helps in understanding the relationship between investments and sales and provides insights into future sales trends.


````python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Sample data - yearly sales and investments
years = np.array([1, 2, 3, 4, 5])  # Years
sales = np.array([10000, 20000, 30000, 40000, 50000])  # Total sales for each year
investments = np.array([5000, 6000, 7000, 8000, 9000])  # Total investments for each year

# Define the deep learning model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=[1]),  # Input layer
    tf.keras.layers.Dense(units=64, activation='relu'),  # Hidden layer
    tf.keras.layers.Dense(units=1)  # Output layer
])

# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")

# Train the model
model.fit(years, sales, epochs=1000, verbose=0)  # Training the model on sales data

# Predict annual sales based on investments
predicted_sales = model.predict(years)

# Calculate profit
profit = predicted_sales.flatten() - investments

# Visualize the results
plt.figure(figsize=(10, 6))

# Plot actual sales and predicted sales
plt.bar(years - 0.2, sales, width=0.4, color='b', align='center', label='Actual Sales')
plt.bar(years + 0.2, predicted_sales.flatten(), width=0.4, color='r', align='center', label='Predicted Sales')

# Plot investments as a reference
plt.plot(years, investments, marker='o', linestyle='-', color='g', label='Investments')

# Plot profit
plt.bar(years, profit, width=0.4, color='y', align='center', label='Profit')

# Predict sales for future years
future_years = np.array([6, 7, 8, 9, 10])
predicted_sales_future = model.predict(future_years)
plt.bar(future_years, predicted_sales_future.flatten(), width=0.4, color='m', align='center', label='Predicted Sales (Future)')

plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Annual Sales vs Investments')
plt.xticks(np.concatenate([years, future_years]))  # Include future years on x-axis
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
````

# Deep Learning Model for Predicting Annual Sales

This code snippet demonstrates the use of a deep learning model to predict annual sales based on investment data.

## Sample Data Description

- **Years**: An array representing the years over the given period. For example, `[1, 2, 3, 4, 5]`.
- **Sales**: Total sales for each year. For example, `[10000, 20000, 30000, 40000, 50000]`.
- **Investments**: Total investments made for each year. For example, `[5000, 6000, 7000, 8000, 9000]`.

## Deep Learning Model

The deep learning model is defined using TensorFlow's Keras API:
1. **Input Layer**: Dense layer with 64 units and ReLU activation function.
2. **Hidden Layer**: Dense layer with 64 units and ReLU activation function.
3. **Output Layer**: Dense layer with 1 unit.

The model is compiled using the Adam optimizer and mean squared error loss function.

## Training

The model is trained using the `fit()` function for 1000 epochs with sales data as the target variable.

## Predictions

- Annual sales are predicted based on the provided investment data.
- Profit is calculated by subtracting investments from predicted sales.

## Results

Here are the profits calculated for each year based on the predicted sales and actual investments:

- Profit for year 1: $2821.46
- Profit for year 2: $7435.41
- Profit for year 3: $10995.22
- Profit for year 4: $12976.48
- Profit for year 5: $14659.98

These profits represent the difference between predicted sales and actual investments for each year.

This analysis helps in understanding the potential profitability of sales based on investment decisions.


````python
import numpy as np
import tensorflow as tf

# Sample data - yearly sales and investments
years = np.array([1, 2, 3, 4, 5])  # Years
sales = np.array([10000, 20000, 30000, 40000, 50000])  # Total sales for each year
investments = np.array([5000, 6000, 7000, 8000, 9000])  # Total investments for each year

# Define the deep learning model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=[1]),  # Input layer
    tf.keras.layers.Dense(units=64, activation='relu'),  # Hidden layer
    tf.keras.layers.Dense(units=1)  # Output layer
])

# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")

# Train the model
model.fit(years, sales, epochs=1000, verbose=0)  # Training the model on sales data

# Predict annual sales based on investments
predicted_sales = model.predict(years)

# Calculate profit
profit = predicted_sales.flatten() - investments

# Print profit for each year
for year, profit_value in zip(years, profit):
    print(f"Profit for year {year}: ${profit_value:.2f}")
````
